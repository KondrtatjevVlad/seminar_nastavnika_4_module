```scala
%spark.pyspark

from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```scala
%spark.pyspark

users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"]
)

orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1", 2, 10.0),
        ("o2", "u1", "p2", 1, 30.0),
        ("o3", "u2", "p1", 1, 10.0),
        ("o4", "u2", "p3", 5, 7.0),
        ("o5", "u3", "p2", 3, 30.0),
        ("o6", "u3", "p3", 1, 7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"]
)

products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"]
)

users.show()
orders.show()
products.show()
```

```scala
%spark.pyspark

orders_with_revenue = orders.withColumn(
    "revenue",
    F.col("qty") * F.col("price")
)

orders_with_revenue.show()
```

```scala
%spark.pyspark

orders_enriched = (
    orders_with_revenue
    .join(users, on="user_id", how="inner")
    .join(products, on="product_id", how="inner")
)

orders_enriched.show()
```

```scala
%spark.pyspark

city_product_metrics = (
    orders_enriched
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.count("order_id").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum")
    )
)

city_product_metrics.show()
```

```scala
%spark.pyspark

window_city = Window.partitionBy("city").orderBy(F.desc("revenue_sum"))

mart_city_top_products = (
    city_product_metrics
    .withColumn(
        "rank",
        F.row_number().over(window_city)
    )
    .filter(F.col("rank") <= 2)
    .select(
        "city",
        "rank",
        "product_id",
        "product_name",
        "orders_cnt",
        "qty_sum",
        "revenue_sum"
    )
)

mart_city_top_products.orderBy("city", "rank").show()
```

```scala
%spark.pyspark

hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"

mart_city_top_products.write.mode("overwrite").parquet(hdfs_path)
```

```scala
%spark.pyspark

result_from_hdfs = spark.read.parquet(hdfs_path)

result_from_hdfs.orderBy("city", "rank").show()
```
